Для анализа рынка были учтены только квартиры на долгий срок и три вида квартир: однокомнатные, двухкомнатные и трехкомнатные квартиры соответственно. ### **Результаты:**

В результате анализа было выяснено, что в Астане в этот день сдавалось 412 однокомнатных, 748 двухкомнатных и 435 трехкомнатных квартир. А в Алматы сдавалось намного больше: 1109 однокомнатных, 2325 двухкомнатных и 1176 трехномнатных квартир.

Медианная цена для квартиры в Астане - 360 000 тенге, в Алматы - 300 000.

Чем больше комнат, тем меньше цена за кв. метр:
Цена на квадратный метр однокомнатной квартиры в Астане составил 5208 тг., а в Алматы на *28% больше* - 6667 тг. Квадратный метр трехкомнатной в Астане стоил 4396 тг., в Алматы 5608 тг.

В Астане средняя цена для

1-комнатной - ₸200 000.

2-комнатной - ₸300 000

3-комнатной - ₸450 000.

В Алматы средняя цена для

1-комнатной - ₸250 000.

2-комнатной - ₸380 000.

3-комнатной - ₸500 000.

Несмотря на значительный отрыв в количестве квартир в Алматы, там квартиры *дороже в среднем на 20%* чем в Астане. Возможно на цены в Алматы сказывается большое количество студентов в городе.

Для сравнения:
В 2023 году в Астане за квадратный метр просили в среднем 4815 тг., а в г. Алматы - 5549 тг. Это цены в среднем за год, без учета сезонности.


### **А как было в прошлых годах?**

Данные из stat.gov дают шанс взглянуть на состояние рынка с 2001 года по всей Республике:
1. Хаотичное формирование рынка аренды жилья до 2006-года;
2. Кризис 2008-года вероятно повлиял на снижение цены почти во всех городах;
3. Заметное подорожание видно в отрезке 2012-2013 годов - как отражение восстановления рынка после глобального кризиса. [Возврат инвесторского интереса и активизация на рынке жилья сопровождались восстановлением строительства после спада 2009 года.](https://colab.research.google.com/drive/1wQmtgpdE-12ikfFZY-XZr-V1B3cTr2XD#scrollTo=cDz1fjRdKUu7&line=42&uniqifier=1);
4. Снижение цен на сырье на мировых рынках и "отправление в свободное плавание" тенге в 2015 видимо не повлиял на рынок жилья;
5. В "ковидное" время наблюдаем подорожание;
6. А на мобилизацию в РФ в 2022-году рынок ответил резким подорожанием аренды жилья. Событие стало таким значимым, что спустя два года цены не сильно снизились.

### Подключаемся к sqlite3 и экспортируем наши данные в отдельные csv файлы (в итоге получаем 2 штуки) и соединяем их в один:

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('db.sqlite')

query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql_query(query, conn)
print("Tables:", tables)

for table_name in tables['name']:
    df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
    df.to_csv(f"{table_name}.csv", index=False)
    print(f"Exported {table_name} to {table_name}.csv")

conn.close()

from google.colab import files
for table_name in tables['name']:
    files.download(f"{table_name}.csv")
```

### Определяем выбросы с помощью Z-теста и избавляемся от них:

С помощью библиотеки plotly выводим следующие графики:

1. Средняя цена квартир по городам

2. Распределение цен по городам

3. Средняя цена квартир (по количеству комнат) по городам

4. Средняя цена квадратного метра по городам

5. Распределение цен квартир по ценам и кв. метру.

```python
import plotly.express as px

overall_clean = pd.read_csv('overall_clean3.csv')
avg_price_by_city = overall.groupby('city')['price'].mean().reset_index()

fig = px.bar(avg_price_by_city, x='city', y='price', title='Средняя цена по городам', labels={'price':'Средняя цена', 'city': 'Город'})
fig.show()

fig = px.box(overall_clean, x='city', y='price', title='Распределение цен по городам', labels={'price':'Цена', 'city': 'Город'})
fig.show()

avg_price_room_city = overall_clean.groupby(['city', 'room'])['price'].median().reset_index()

fig = px.bar(avg_price_room_city, x='room', y='price', color='city', barmode='group',
             title='Средняя цена квартир (по количеству комнат) по городам',
             labels={'price': 'Средняя цена', 'room': 'Кол-во комнат', 'city': 'Город'})
fig.show()

overall_clean['price_per_sqm'] = overall_clean['price'] / overall_clean['square']
price_per_sqm_by_city = overall_clean.groupby(['city', 'room'])['price_per_sqm'].median().reset_index()

fig = px.bar(price_per_sqm_by_city, x='room', y='price_per_sqm', color='city', barmode='group', title='Средняя цена кв. метра по городам', labels={'price_per_sqm':'Цена кв. метра', 'city': 'Город'})
fig.show()

fig = px.scatter(overall_clean, x='square', y='price', color='city',
                 title='Распределение цен квартир по ценам и кв. метру',
                 labels={'square':'кв. метр', 'price':'Цена', 'city': 'Город'})
fig.show()

fig = px.histogram(overall_clean, x='room', color='city', barmode='group',
                   title='Количество квартир (по числу комнат) в городах',
                   labels={'room':'Количество квартир по числу комнат', 'count': 'Количество', 'city': 'Город'})
fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=3, cols=2,
    subplot_titles=(
        'Средняя цена в городах', 'Распределение количества комнат по городам',
        'Средняя цена квартир (по количеству комнат) по городам', 'Средняя цена квадратного метра по городам',
        'Распределение цен по городам', 'Распределение цен квартир по ценам и кв. метру.'
    ),
    vertical_spacing=0.2, horizontal_spacing=0.1
)

avg_price_by_city = overall_clean.groupby('city').agg(avg_price=('price', 'median')).reset_index()
fig.add_trace(
    go.Bar(x=avg_price_by_city['city'], y=avg_price_by_city['avg_price'], name='Average Price by City'),
    row=1, col=1
)

room_count_by_city = overall_clean.groupby('city')['room'].value_counts().unstack().fillna(0)
for room in room_count_by_city.columns:
    fig.add_trace(
        go.Bar(x=room_count_by_city.index, y=room_count_by_city[room], name=f'{room}-Room Count'),
        row=1, col=2
    )

avg_price_room_city = overall_clean.groupby(['city', 'room']).agg(avg_price=('price', 'median')).reset_index()
for city in avg_price_room_city['city'].unique():
    city_data = avg_price_room_city[avg_price_room_city['city'] == city]
    fig.add_trace(
        go.Bar(x=city_data['room'], y=city_data['avg_price'], name=f'Средняя цена в {city}', legendgroup=city),
        row=2, col=1
    )

price_per_sqm_by_city = overall_clean.groupby(['city', 'room']).agg(price_per_sqm=('price_per_sqm', 'median')).reset_index()
for city in price_per_sqm_by_city['city'].unique():
    city_data = price_per_sqm_by_city[price_per_sqm_by_city['city'] == city]
    fig.add_trace(
        go.Bar(x=city_data['room'], y=city_data['price_per_sqm'], name=f'Цена/кв. метр в {city}', legendgroup=city),
        row=2, col=2
    )

for city in overall_clean['city'].unique():
    city_data = overall_clean[overall_clean['city'] == city]
    fig.add_trace(
        go.Box(y=city_data['price'], name=city, boxpoints='outliers', marker_color='rgba(255,0,0,0.7)'),
        row=3, col=1
    )

for city in overall_clean['city'].unique():
    city_data = overall_clean[overall_clean['city'] == city]
    fig.add_trace(
        go.Scatter(x=city_data['square'], y=city_data['price'], mode='markers', name=f'{city}', marker=dict(size=8)),
        row=3, col=2
    )

fig.update_layout(
    height=1000, width=1200,
    title_text="Обзор рынка аренды квартир",
    showlegend=False,
    template="plotly_white"
)

fig.show()
```

Выводим:
1. график изменения цен по годам в регионах
2. сравнение цен 2001 и 2023 годов с помощью plotly

```python
df_melted = pd.melt(df, id_vars=['city'], var_name='year', value_name='price')

df_melted['year'] = pd.to_numeric(df_melted['year'])

fig = px.line(df_melted, x='year', y='price', color='city',
              title='График цен в регионах (2001-2023)',
              labels={'city': 'Город', 'price': 'Цена аренды', 'year': 'Год'})
fig.show()

df_2001_2023 = df_melted[df_melted['year'].isin([2001, 2023])]
fig = px.bar(df_2001_2023, x='city', y='price', color='year',
             barmode='group', title='Сравнение цен 2001 и 2023',
             labels={'price': 'Цена аренды', 'city': 'Город', 'year': 'Год'})
fig.show()
```

```python
df_melted['price_pct_change'] = df_melted.groupby('city')['price'].pct_change() * 100

fig = px.line(df_melted, x='year', y='price_pct_change', color='city',
              title='Темп роста цен относительно прошлого года (%)',
              labels={'price_pct_change': 'Рост цены (%)', 'year': 'Год', 'city': 'Город'})
fig.show()
```
